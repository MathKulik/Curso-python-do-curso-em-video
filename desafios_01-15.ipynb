```python
# exercicio 004

a = input('Digite algo: ')
print('O tipo desse valor é ', type(a))
print('Só tem espaçõ? ', a.isspace())
print('É numero? ', a.isnumeric())
print('É alfabetico?', a.isalpha())
print('É alfanumerico? ', a.isalnum())
print('Tá maiuscula? ', a.isupper())
print('Tá minuscula? ', a.islower())
print('Tá capitalizada? ', a.istitle())
```
